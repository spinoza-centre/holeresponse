```python
%matplotlib inline
subj_list = subj_obj.get_subjects()
ncols = 3
nrows = int(np.ceil(len(subj_list)/ncols))

fig,axs = plt.subplots(
    ncols=ncols, 
    nrows=nrows,
    figsize=(ncols*4,nrows*4), 
    constrained_layout=True
)

for ix,ax in enumerate(axs.flatten()):

    if (ix+1)>len(subj_list):
        # remove obsolete axes
        ax.remove()
    else:
        sub = subj_list[ix]
        print(f"plotting '{sub}'")

        if sub == "sub-010":
            ses = 5
        else:
            ses = None

        tmp_obj = hr.data.T2StarSlices(sub, ses=ses)
        tmp_obj.plot_intensity(
            # ribbon=(350,370),
            axs=ax,
            normalize=True,
            # avg=True,
            plot_kw={
                "title": {
                    "title": tmp_obj.subject,
                    "fontweight": "bold"
                },
            }
        )

fig.suptitle("T$_2$*-intensity across line", fontsize=30, y=1.05)
# hr.viz._save_figure(fig, fname="group_desc-composite_stims", overwrite=True)
```

```python
%matplotlib inline
subj_list = subj_obj.get_subjects()
ncols = 3
nrows = int(np.ceil(len(subj_list)/ncols))

normalize = True
fig,axs = plt.subplots(
    ncols=ncols, 
    nrows=nrows,
    figsize=(ncols*4,nrows*4), 
    constrained_layout=True,
    sharey=normalize
)

for ix,ax in enumerate(axs.flatten()):

    if (ix+1)>len(subj_list):
        # remove obsolete axes
        ax.remove()
    else:
        sub = subj_list[ix]
        print(f"plotting '{sub}'")

        if sub == "sub-010":
            ses = 5
        else:
            ses = None

        tmp_obj = hr.data.T2StarSlices(
            sub, 
            ses=ses,
            ribbon=(),
        )

        tmp_obj.plot_intensity(
            axs=ax,
            normalize=True,
            # avg=True,
            plot_kw={
                "title": {
                    "title": tmp_obj.subject,
                    "fontweight": "bold"
                },
            }
        )

fig.supxlabel("time (s)", fontsize=24)
fig.supylabel("norm(T$_2$*)", fontsize=24)
fig.suptitle("T$_2$*-intensity across depth", fontsize=30, y=1.05)
# hr.viz._save_figure(fig, fname="group_desc-composite_stims", overwrite=True)
```
